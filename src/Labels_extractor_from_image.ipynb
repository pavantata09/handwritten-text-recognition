```python
DIR = '/data2/pavan/Data/crop'
DIR
```

```python
image_DIR = "/data2/pavan/Data/"
if not os.path.exists(DIR):
    os.mkdir(DIR)
filename = ''
img = np.array([])
croppedSave_DIR = DIR
for index, rows in df.iterrows():
    #print(rows)
    if filename !=rows['filename']:
        count = 0;
        filename = rows['filename']
    label = json.loads(rows['region_attributes'])
    if label:
        img = cv2.imread('{}/{}'.format(image_DIR, filename))
        print(filename)
        print(img.shape)
        #aadhaar_DIR= '{}/'.format(DIR) + folder['bbox']
        if not os.path.exists(croppedSave_DIR):
            os.mkdir(croppedSave_DIR)
        bbox = json.loads(rows['region_shape_attributes'])
        if bbox['y'] < 0:
            bbox['y'] = 1
        cropped_image = img[bbox['y']:bbox['y']+bbox['height'],bbox['x']:bbox['x'] + bbox['width']]
        print("At index:",index)
        print("croppped_image_shape:", cropped_image.shape)
        count = count + 1
        cv2.imwrite('{}/{}-{}'.format(croppedSave_DIR, index,filename),cropped_image)
        with open('{}/train.txt'.format(DIR), 'a') as f:
            f.writelines('/data2/pavan/handwritten-text-recognition/src/images/{}-{} {}\n'.format(index,filename, label['bbox']))
```
